## Create charts showing demographics of middle TN

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#import NDC deliveries csv and skip rows at the end
county_stats = pd.read_csv('../data/Nash Diaper Stastistics - Population Breakdown by County.csv',
                  skipfooter = 12,  
                  engine = 'python')
county_stats.head()
```

```python
#rename columns
county_stats = county_stats.rename(columns = {'Total Population 1':'Total Population',
                                              'Est # in Diapers 5':'Est # in Diapers',
                                              'Est. TennCare 4':'Est. TennCare',
                                              'Household Income (Avg) 2':'Household Income (Avg)',
                                              '% of Title 1 Schools':'% Title 1 Schools',
                                              'Food Stamp % 3':'% Food Stamp',
                                              'Poverty Rate 0-17 2':'% Poverty Rate 0-17',
                                              'Poverty Rate All Ages 2':'% Poverty Rate All Ages'
                                             })
county_stats.columns
```

```python
#strip % from certain columns
county_stats['% Title 1 Schools'] = county_stats['% Title 1 Schools'].str.rstrip('%')
county_stats['% Food Stamp'] = county_stats['% Food Stamp'].str.rstrip('%')
county_stats['% Poverty Rate 0-17'] = county_stats['% Poverty Rate 0-17'].str.rstrip('%')
county_stats['% Poverty Rate All Ages'] = county_stats['% Poverty Rate All Ages'].str.rstrip('%')
county_stats.head()
```

```python
#create grouped bar chart showing poverty in counties
tn_poverty = sns.load_dataset("county_stats")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=county_stats, kind="bar",
    x="County", y="body_mass_g", hue="sex",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
```
