```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
#import NDC deliveries csv
county_stats = pd.read_csv('../files/county_stats.csv')
county_stats.head()
```

```python
#only keep needed columns
county_stats = county_stats[['County',
                             'Population Age 0-4',
                             'Est # in Diapers',
                             'Est. TennCare',
                             'In Diapers in Poverty',
                             'diapers used'
                            ]].copy()
county_stats.head()
```

```python
#rename columns
county_stats = county_stats.rename(columns = {'County':'county',
                                              'Population Age 0-4':'pop_under_4',
                                              'Est # in Diapers':'est_in_diapers',
                                              'Est. TennCare':'est_in_tenncare',
                                              'In Diapers in Poverty':'diapers_and_poverty',
                                              'diapers used':'diapers_used'
                                             })
county_stats.columns
```

```python
#import NDC deliveries csv
total_by_county = pd.read_csv('../files/total_by_county.csv')
total_by_county.head()
```

```python
#strip TN from counties and rename
total_by_county['County'] = total_by_county['County'].str.strip(', TN')
total_by_county = total_by_county.rename(columns = {'County':'county',
                                              'Total':'total'
                                             })
total_by_county
```

```python
#convert total column from year to month
total_by_county['total']=total_by_county['total']/12
total_by_county
```

```python
#merge datasets
diapers_by_county = pd.merge(county_stats, total_by_county, on = 'county', how = 'outer')
diapers_by_county
```

```python
#move total to correct spot
diapers_by_county.loc[10, 'total'] = diapers_by_county.loc[12, 'total']
#replace null values with zero
diapers_by_county['total'] = diapers_by_county['total'].fillna(0)
#delete last 2 rows
#diapers_by_county = diapers_by_county.drop(['otal'])#df.drop(['Cochice', 'Pima'])
diapers_by_county
```

```python
#delete last 2 rows
diapers_by_county = diapers_by_county.drop(diapers_by_county.index[[11,12]])
diapers_by_county
```

```python
diapers_by_county.info()
```

```python
#convert to integers
diapers_by_county['diapers_and_poverty'] = diapers_by_county['diapers_and_poverty'].astype(int)
diapers_by_county['diapers_used'] = diapers_by_county['diapers_used'].astype(int)
diapers_by_county['total'] = diapers_by_county['total'].astype(int)
```

```python
#add 'diaper_need' column
diapers_by_county['diapers_need'] = diapers_by_county['diapers_used']-diapers_by_county['total']
diapers_by_county
```

```python
# Setting the positions and width for the bars
pos = list(range(len(diapers_by_county['diapers_used']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        diapers_by_county['diapers_used'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=diapers_by_county['county'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        diapers_by_county['diapers_used'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=diapers_by_county['county'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        diapers_by_county['diapers_need'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=diapers_by_county['county'][2]) 

# Set the y axis label
ax.set_ylabel('Number of Diapers')

# Set the chart's title
ax.set_title('County')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(diapers_by_county['county'])

# Setting the x-axis and y-axis limits
#plt.xlim(min(pos)-width, max(pos)+width*4)
#plt.ylim([0, max(diapers_by_county['diapers_used'] + diapers_by_county['total'] + diapers_by_county['diapers_need'])] )

# Adding the legend and showing the plot
plt.legend(['Used per Month', 'Donated by NDC', 'Need Remaining'], loc='upper left')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))

# set width of bar
barWidth = 0.25
 
# set height of bar
bars1 = diapers_by_county['diapers_used']
bars2 = diapers_by_county['total']
bars3 = diapers_by_county['diapers_used']
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='#ffac00', width=barWidth, edgecolor='white', label='Used per Month')
plt.bar(r2, bars2, color='#985ca3', width=barWidth, edgecolor='white', label='Donated by NDC')
plt.bar(r3, bars3, color='#b1d863', width=barWidth, edgecolor='white', label='Remaining Need')
 
# Add xticks on the middle of the group bars
plt.xlabel('County', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], diapers_by_county['county'], rotation=90)

plt.ylabel('Number of Diapers', fontweight='bold')

plt.title('Diaper Need in Middle TN by County')
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
