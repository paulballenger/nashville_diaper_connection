## Plot poverty in TN by year

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
#import hstpov21 csv
hist_poverty = pd.read_csv('../data/hstpov21.csv', header=4)
hist_poverty.head()
```

```python
#only keep rows for TN
keep_list = ['Tennessee']
hist_poverty=hist_poverty[hist_poverty['STATE'].isin(keep_list)].reset_index()
hist_poverty = hist_poverty.drop(columns=['index'])
hist_poverty.head(30)
```

```python
#delete row with duplicate 2013 and 2017 data
hist_poverty = hist_poverty.drop(hist_poverty.index[[2,7]])
hist_poverty.head(10)
```

```python
#remove info in parentheses from years and drop years with NaN values
hist_poverty = hist_poverty.dropna()
hist_poverty['Year'] = hist_poverty['Year'].str.split().str[0]
hist_poverty.head(25)
```

```python
hist_poverty.info()
```

```python
#for total and number columns - remove commas, convert to integer and multiply by 1000
hist_poverty['Total'] = hist_poverty['Total'].replace(',','', regex=True)
hist_poverty['Total'] = hist_poverty['Total'].astype(int)*1000
hist_poverty['Below poverty'] = hist_poverty['Below poverty'].replace(',','', regex=True)
hist_poverty['Below poverty'] = hist_poverty['Below poverty'].astype(int)*1000
hist_poverty.head()
```

```python
#export as csv
hist_poverty.to_csv('../data/hist_poverty.csv', index = False)
```
